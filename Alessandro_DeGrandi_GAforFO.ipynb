### Genetic Algorithm Class

```
# It's a class that implements a genetic algorithm for the De Jong function
class GeneticAlgorithmForDJ:

  def __init__(self, dj_instance, selection_rule, elitism=False, population_array=None, population_size=20,
               mutation_rate=0.01, iterations=600):
    """
    This function initializes the genetic algorithm with the given parameters.

    :param dj_instance:      This is the instance of the De Jong model that you want to optimise.
    :param selection_rule:   The selection rule to be used. Can be either 'rank' or 'roulette'.
    :param elitism:          True or False. If True, the best individual in the population will be carried over
                             to the next generation. Defaults to False (optional), in this case only the population
                             is replaced by the new generation.
    :param population_array: Optional parameter, can contain an initial array with the population. It is a list of
                             lists, where each inner list correspond to an individual of the population.
    :param population_size:  The number of individuals in the population, defaults to 20 (optional).
    :param mutation_rate:    The probability of a gene mutating, defaults 0.01 (optional).
    :param iterations:       The number of iterations to run the algorithm for, defaults to 600 (optional).
    """

    self.dj_instance = dj_instance
    self.dimension = self.dj_instance.dimension
    self.dimension_bits = self.dj_instance.num_bits
    self.selection_rule = selection_rule
    self.population_size = population_size
    self.population_array = population_array
    self.mutation_rate = mutation_rate
    self.iterations = iterations
    self.elitism = elitism

    # Encoding functions
    self.encoding_function = np.vectorize(self.dj_instance.gray_encode)
    self.decoding_function = np.vectorize(self.dj_instance.gray_decode)

    # Statistics variables
    self.best_fitness = []
    self.best_individuals = []
    #self.mean_fitness = []

    # Additional variables
    self.population = None
    self.fitness = None

    self.old_population=None
    self.old_fitness=None

    self.parents = None

    self.best_solution=None
    self.best_solution_fitness = None
    self.best_solution_iteration = 0
    #self.mean_fitness_value = None


  def solve(self):
    """
    This function is the main of the class `GeneticAlgorithm`. It performs the main loop of the algorithm,
    and it calls the other functions of the class.
    """
    # Initialise population
    self.initialise_population()

    # Evaluate population
    self.compute_fitness()
    self.best_solution_fitness=max(self.fitness)

    for iteration in range(self.iterations):

      self.old_population = self.population
      self.old_fitness = self.fitness

      # Select parents
      self.selection()

      # Generates new individuals using reproduction rules
      self.crossover()
      self.mutation()

      # Form a new population
      if(self.elitism):
          self.elitism_selection()

      # Evaluate population
      self.compute_fitness()

      self.best_fitness.append(max(self.fitness))
      self.best_individuals.append(self.population[np.argmax(self.fitness)])
      if(max(self.fitness)>self.best_solution_fitness):
          self.best_solution_fitness=max(self.fitness)
          self.best_solution = self.population[np.argmax(self.fitness)]
          self.best_solution_iteration = iteration


  def initialise_population(self):
    """
    This function generates a random population of shape `population_size` and `dimension` with values between
    `range[0]` and `range[1]`.
    :return: The population encoded using gray code.
    """

    self.population = self.encoding_function(np.random.uniform(self.dj_instance.range[0],self.dj_instance.range[1],(self.population_size,self.dimension)))
    #print(self.population)

  def compute_fitness(self):
    """
    :return:
    """

    self.fitness = self.dj_instance.evaluate(self.decoding_function(self.population))


    # Compute the fitness
    #high value of function evaluation needs to have lower fitness, fitness needs to be always positive to get probabilities

    self.fitness = 1 / np.exp(self.fitness) #print("FITNESS",self.fitness)

  def rank_selection(self):
    """
    This function compute the individual probability of selection using rank selection rule.
    :return: The probability of each individual being selected for the next generation.
    """

    self.population = self.population[self.fitness.argsort()][::-1]
    sum_ranks =  (self.population_size-1) * (self.population_size) / 2
    # Compute the probability of selection
    probability = np.zeros(self.population_size)
    for i in range(self.population_size):
      probability[i] = (i)/sum_ranks

    #print("Probabilities rank",probability)
    #print("sum = ",np.sum(probability))
    return probability


  def roulette_selection(self):
    """
    This function compute the individual probability of selection using roulette selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    # Compute the probability of selection
    probability = self.fitness / np.sum(self.fitness)
    #print("Probabilities roulette",probability)
    #print("sum = ",np.sum(probability))

    return probability

  def selection(self):
    """
    This function is used to select the parents for the next generation according to a selection rule.
    """
    # Computing the probability of each individual to be selected as parent.
    if self.selection_rule == 'rank':
        probability = self.rank_selection()
    elif self.selection_rule == 'roulette':
        probability = self.roulette_selection()
    else:
        raise ValueError('The selection rule must be either rank or roulette.')

    # Select the individuals
    parentsIndex = np.random.choice(np.arange(0,self.population_size) , size=self.population_size, p=probability)
    self.parents = self.population[parentsIndex]


  def crossover(self):
      """
      This function perform a single-point crossover on each pair of parents, selected by pairing the first half of
      the parents with the second half of them.
      """

      new_population = self.encoding_function(np.random.uniform(self.dj_instance.range[0],self.dj_instance.range[1],(self.population_size,self.dimension)))

      #print("SIZE",self.parents.size)
      #print("SHAPE",self.parents.shape)
      #print("PARENTS",self.parents)
      # Perform the crossover
      for i in range(0, self.parents.shape[0]//2):
          # Select the parents
          parent1 = self.parents[i]
          parent2 = self.parents[i + self.parents.shape[0]//2]


          # Select the crossover points
          crossover_points = np.random.randint(0, self.dimension_bits, self.dimension)

          #print("PARENT1",parent1)
          #print("PARENT2",parent2)
          #print("cross_index",crossover_points)

          child1=parent1
          child2=parent2
          # Perform the crossover
          for c,crossover_point in enumerate(crossover_points):

            #print(parent1[i])
            #print(parent2[i])

            child1[c] = parent1[c][:crossover_point] + parent2[c][crossover_point:]
            child2[c] = parent2[c][:crossover_point] + parent1[c][crossover_point:]

          #print("CHILD1",child1)
          #print("CHILD2",child2)
          #print(new_population)
          new_population[i] = child1
          new_population[i + 1] = child2
      # Update the population
      self.population = new_population

  def mutation(self):
    """
    This function, decide if mutate or not each bit of each offspring in each child, by flip it, with a mutation
     rate, otherwise keep it the same.
    """
    # Create the new population
    new_population =  self.encoding_function(np.random.uniform(self.dj_instance.range[0],self.dj_instance.range[1],(self.population_size,self.dimension)))
    # Perform the mutation
    for i in range(self.population_size):
        for j in range(self.dimension):
            gene = list(self.population[i][j])
            #print(gene)
            for k in range(self.dimension_bits):
              if np.random.rand() < self.mutation_rate:
                  gene[k] = str(1 - int(self.population[i][j][k]))
            #print(gene)
            new_population[i][j] = ''.join(gene)

    # Update the population
    self.population = new_population

  def elitism_selection(self):
    """
    Generating a new population as combination of the old one and the new one.
    """
    temp_population = np.concatenate((self.old_population,self.population))
    temp_fitness = np.concatenate((self.old_fitness,self.fitness))


    pop_indexes = np.argpartition(temp_fitness,-self.population_size)[-self.population_size:] #indices of n greatest elements
    #print(pop_indexes)

    self.population = temp_population[pop_indexes]
    """
    new_population = population
    for i in range(self.population_size):
      if(pop_indexes[i]<self.population_size):
        new_population[i]= self.old_population[pop_indexes[i]]
      else:
        new_population[i]= self.population[pop_indexes[i]]
    """

    # Update the population
    #self.population=new_population

  #def reconstruction(self):
  #  """
  #  If elitism is enabled, this function replace the population with the 20 best individuals.
  #  """

  def update_statistics(self):
    """
    This function can be used to compute statistics on the objective value.
    """
```

### Settings

```
# population_size = 20
selection_rules = ['rank', 'roulette']
bool_elitism = [False, True]
methods = ['rank', 'rank + elitism', 'roulette', 'roulette + elitism']#['roulette', 'roulette + elitism']
# Single-Point Crossover
mutation_rate = 0.01
iterations = 1000
seeds = [0, 42, 666] # (Optional)

# Represent each individual of your population as a string of bits
# Each individual basically represent a point in you De Jong Function
functions = [DeJong(1), DeJong(2), DeJong(3), DeJong(4), DeJong(5)]
```

#### Test

```
c_labels = ['Function', 'Method', 'Seed', 'Best solution', 'Best evaluation', 'Best fitness value', 'Best iteration']
data = []

for function in functions:
  print(function.function_name)
  # Plotting the results of the different methods for each seed.
  for seed in seeds:
    np.random.seed(seed)
    fig, axes = plt.subplots(nrows=1, ncols=len(methods), figsize=(36, 8))
    fig.suptitle(f'{function.function_name}, Random Seed: {seed}')

    # Creating a new instance of the GeneticAlgorithmForDJ class for each selection rule and elitism combination, running the
    # algorithm for each of the methods and plotting the results.
    for r_idx, sr in enumerate(selection_rules):
      for e in bool_elitism:
        GA = GeneticAlgorithmForDJ(dj_instance=function, selection_rule=sr, elitism=e, iterations=iterations)
        GA.solve()

        best_solution = [function.gray_decode(i) for i in GA.best_solution]
        print(sr+" elitism="+str(e))
        print("best_solution= ",best_solution)
        print("best_evaluation= ",GA.dj_instance.evaluate([best_solution]))
        print()
        # Plot
        axes[r_idx *2  + e].plot(range(len(GA.best_fitness)),GA.best_fitness)

        # Populate dataframe
        data.append([function.function_name,methods[r_idx *2  + e],seed,best_solution, GA.dj_instance.evaluate([best_solution]),GA.best_solution_fitness,GA.best_solution_iteration])

  plt.show()

df = pd.DataFrame(data, columns=c_labels)
```
